Removal of punctuations

```python
string.punctuation
```

```python

def remove_punctuation(text):

    punctuation_chars = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    return ''.join(char for char in text if char not in punctuation_chars)
```

```python
list1=[]
for idx,string in enumerate((df['Review'])):
  list1.append(remove_punctuation(string))
df['preprocessed_review']=list1
```

```python
list01=[]
for idx,string in enumerate((df['First_Expression'])):
  list01.append(remove_punctuation(string))
df['preprocessed_firstexpression']=list01
```

```python
import emoji
import re

def remove_emojis(text):

    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1E0-\U0001F1FF"
                               u"\U00002500-\U00002BEF"
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
list2=[]
for idx,string in enumerate((df['preprocessed_review'])):
  list2.append(remove_emojis(string))
df['preprocessed_emoji']=list2
```

```python
list02=[]
for idx,string in enumerate((df['preprocessed_firstexpression'])):
  list02.append(remove_emojis(string))
df['preprocessed_first_emoji']=list02
```

**Removal of Stopwords using nltk library**

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords
nltk.download('stopwords')
```

```python
nltk.download('punkt')
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    tokens = nltk.word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)
```

```python
list3=[]
for idx,string in enumerate((df['final_preprocessed'])):
  list3.append(remove_stopwords(string))
df['preprocessed_stopwords']=list3
```
